```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
pop = pd.read_excel('CAORAN.xlsx')[pd.read_excel('CAORAN.xlsx')['observation_date'] >= 2002][['observation_date', 'CAORAN7POP']]
pop['CAORAN7POP'] = pop['CAORAN7POP']*1000

plt.plot(pop['observation_date'], pop['CAORAN7POP'])
plt.xlabel('Year')
plt.ylabel('Population')
plt.title('OC Population')
plt.xlim(2000, 2025)
plt.grid()
plt.show()
```

```python
ca_gdp = pd.read_csv('gross domestic product (gdp) state  california (dollars).csv')[
    pd.read_csv('gross domestic product (gdp) state  california (dollars).csv')['Years'] >= 2002].iloc[::-1].reset_index(drop=True)

oc_gdp = pd.read_excel('CAORAN.xlsx')[pd.read_excel('CAORAN.xlsx')['observation_date'] >= 2002].reset_index(drop=True)
```

```python
oc_gdp = oc_gdp.reset_index(drop=True)
oc_gdp['GDP'] = ca_gdp['California']/ ((ca_gdp['California'] / oc_gdp['GDPALL06059']).mean())
```

```python
#plt.plot(oc_gdp['observation_date'], oc_gdp['GDPALL06059'])
plt.plot(oc_gdp['observation_date'], oc_gdp['GDP'])
#plt.plot(ca_gdp['Years'], ca_gdp['California'])
plt.xlabel('Year')
plt.ylabel('GDP')
plt.title('OC GDP')
plt.xlim(2000, 2025)
plt.grid()
plt.show()
```

```python
VHT = pd.read_excel('d12_select_fwy_VHT.xlsx').groupby(
    ['Year', pd.read_excel('d12_select_fwy_VHT.xlsx')['Freeway'].str[:-2]])['Total VHT'].mean().reset_index()
VMT = pd.read_excel('d12_select_fwy_VMT.xlsx').groupby(
    ['Year', pd.read_excel('d12_select_fwy_VMT.xlsx')['Freeway'].str[:-2]])['Total VMT'].mean().reset_index()
```

```python
plt.plot(VHT[VHT['Freeway']=='SR57']['Year'], VHT[VHT['Freeway']=='SR57']['Total VHT'])
plt.xlabel('Year')
plt.ylabel('Total VHT')
plt.title('OC SR57 VHT')
plt.xlim(2000, 2025)
plt.grid()
plt.show()

plt.plot(VMT[VMT['Freeway']=='SR57']['Year'], VMT[VMT['Freeway']=='SR57']['Total VMT'])
plt.xlabel('Year')
plt.ylabel('Total VMT')
plt.title('OC SR57 VMT')
plt.xlim(2000, 2025)
plt.grid()
plt.show()
```

```python
miles = pd.read_excel('OC_CalTrans_hwy_road_data.xlsx')[pd.read_excel('OC_CalTrans_hwy_road_data.xlsx')['Year'] >= 2002]

plt.plot(miles['Year'], miles['Highway Miles'])
plt.xlabel('Year')
plt.ylabel('Highway Miles')
plt.title('OC Highway Miles')
plt.xlim(2000, 2025)
plt.grid()
plt.show()
```

```python
miles['Miles % Change'] = (miles['Highway Miles'] / miles.loc[miles['Year'] == 2002, 'Highway Miles'].values[0] - 1)
pop['Pop % Change'] = (pop['CAORAN7POP'] / pop.loc[pop['observation_date'] == 2002, 'CAORAN7POP'].values[0] - 1)
oc_gdp['GDP % Change'] = (oc_gdp['GDP'] / oc_gdp.loc[oc_gdp['observation_date'] == 2002, 'GDP'].values[0] - 1)
```

```python
VHT_I5 = VHT[VHT['Freeway'] == 'I5']
VHT_I5['VHT % Change'] = (VHT_I5['Total VHT'] / VHT_I5.loc[VHT_I5['Year'] == 2002, 'Total VHT'].values[0] - 1)
VMT_I5 = VMT[VMT['Freeway'] == 'I5']
VMT_I5['VMT % Change'] = (VMT_I5['Total VMT'] / VMT_I5.loc[VMT_I5['Year'] == 2002, 'Total VMT'].values[0] - 1)

VHT_SR55 = VHT[VHT['Freeway'] == 'SR55']
VHT_SR55['VHT % Change'] = (VHT_SR55['Total VHT'] / VHT_SR55.loc[VHT_SR55['Year'] == 2002, 'Total VHT'].values[0] - 1)
VMT_SR55 = VMT[VMT['Freeway'] == 'SR55']
VMT_SR55['VMT % Change'] = (VMT_SR55['Total VMT'] / VMT_SR55.loc[VMT_SR55['Year'] == 2002, 'Total VMT'].values[0] - 1)

VHT_SR57 = VHT[VHT['Freeway'] == 'SR57']
VHT_SR57['VHT % Change'] = (VHT_SR57['Total VHT'] / VHT_SR57.loc[VHT_SR57['Year'] == 2002, 'Total VHT'].values[0] - 1)
VMT_SR57 = VMT[VMT['Freeway'] == 'SR57']
VMT_SR57['VMT % Change'] = (VMT_SR57['Total VMT'] / VMT_SR57.loc[VMT_SR57['Year'] == 2002, 'Total VMT'].values[0] - 1)
```

```python
plt.figure(figsize=(10, 5))
plt.plot(pop['observation_date'], pop['Pop % Change'], label='Population')
plt.plot(VHT_I5['Year'], VHT_I5['VHT % Change'], label='I5 VHT')
plt.plot(VMT_I5['Year'], VMT_I5['VMT % Change'], label='I5 VMT')
plt.plot(miles['Year'], miles['Miles % Change'], label='Miles')
plt.plot(oc_gdp['observation_date'], oc_gdp['GDP % Change'], label='GDP')

plt.xlabel('Year')
plt.ylabel('Percentage Change')
plt.title('Orange County I5')
plt.legend()
plt.ylim(-0.25, 0.4)
plt.grid()
plt.minorticks_on()
plt.axhline(y=0, color='black', linestyle='--')
plt.show()
```
