```python
file_names = ['d12_i5_VMT2002.xlsx', 'd12_i5_VMT2005.xlsx', 'd12_i5_VMT2010.xlsx', 'd12_i5_VMT2015.xlsx', 'd12_i5_VMT2020.xlsx', 'd12_i5_VMT2022.xlsx']
vmt_df = {}

for file_name in file_names:
    key = file_name.replace('.xlsx', '').replace('d12_i5_', '')
    vmt_df[key] = pd.read_excel(file_name)
```

```python
freeway = 'I405-S'
result_df = pd.DataFrame(columns=['Year', 'Value'])

for key in vmt_df.keys():
    year = int(key[3:])  # Extract the year from the key
    value = vmt_df[key].loc[vmt_df[key]['Freeway'] == freeway, 'Vehicle Miles Traveled (VMT)'].values[0]
    
    # Append a new row to the result DataFrame
    result_df = result_df.append({'Year': year, 'VMT': value}, ignore_index=True)

plt.plot(result_df['Year'], result_df['VMT'])
plt.xlabel('Year')
plt.ylabel('Vehicle Miles Traveled (VMT)')
plt.title('District 12: '+ freeway)
plt.xticks([2002, 2005, 2010, 2015, 2020, 2022])

plt.show()

# Print the resulting DataFrame
#print(result_df)
```
