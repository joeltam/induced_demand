```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
c = np.linspace(0, 5)
m_values = [5]
n_values = [1, 3, 5]
```

```python
plt.figure(figsize=(6, 6))
for n in n_values:
    for m in m_values:
        delta = (c**n * (m/n)**(n/(n+m)) + c**-m * (m/n)**(-m/(n+m))) / ((m/n)**(n/(n+m)) + (m/n)**(-m/(n+m)))
        label = label='n = {}, m = {}'.format(n, m)
        plt.plot(c, delta, label=label)

plt.xlabel(' (Design Cost)')
plt.ylabel(' (Actual Cost)')
plt.title('Sensitivity of n and m')
plt.ylim(0,5)
plt.legend()
plt.grid(True)
plt.show()
```
